```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# The file path is provided by Kaggle in the input directory
file_path = 'ORIGNAL copy.csv'
df = pd.read_csv(file_path, sep=',')

# Display the first 5 rows and basic info
print("First 5 rows of the dataset:")
display(df.head())

print("\nDataset Information:")
df.info()

# Drop the ID column as it's just an identifier
df = df.drop('id', axis=1)

# Check the distribution of the target variable ('cardio')
# 0 = No Cardiovascular Disease, 1 = Has Cardiovascular Disease
print("\nTarget Variable Distribution:")
print(df['cardio'].value_counts())

# Visualize the distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='cardio', data=df)
plt.title('Distribution of Cardiovascular Disease')
plt.show()

print(f"Original dataset shape: {df.shape}")

# 1. Remove entries where diastolic pressure is higher than systolic pressure
df.drop(df[df['ap_hi'] < df['ap_lo']].index, inplace=True)

# 2. Remove entries with highly improbable BP values.
#    We'll use quantile-based filtering to remove the most extreme outliers,
#    which are very likely data entry errors.
df.drop(df[df['ap_hi'] > df['ap_hi'].quantile(0.999)].index, inplace=True)
df.drop(df[df['ap_hi'] < df['ap_hi'].quantile(0.001)].index, inplace=True)

df.drop(df[df['ap_lo'] > df['ap_lo'].quantile(0.999)].index, inplace=True)
df.drop(df[df['ap_lo'] < df['ap_lo'].quantile(0.001)].index, inplace=True)

print(f"New dataset shape after cleaning: {df.shape}")

# Let's look at the stats for BP columns now to see the change
print("\nBlood Pressure stats after cleaning:")
display(df[['ap_hi', 'ap_lo']].describe())


display(df.describe())

# Convert age if stored in days (some Kaggle datasets store age in days)
# Check if 'age' column exists, is not already int64, and looks like days
if 'age' in df.columns and df['age'].dtype != 'int64' and df['age'].max() > 1000:
    # if age stored in days -> convert to years
    df['age_years'] = (df['age'] / 365).round().astype(int)
    df.drop(columns=['age'], inplace=True)
    df.rename(columns={'age_years':'age'}, inplace=True)


# 1. Feature engineering

# If age still in days convert to years (safety check)
if 'age' in df.columns and df['age'].max() > 1000:
    df['age'] = (df['age'] / 365).round().astype(int)

# Map gender to 0/1 (female=0, male=1) if necessary
if 'gender' in df.columns:
    # Assuming Kaggle dataset standard: 1=female, 2=male
    if df['gender'].max() > 1:
        df['gender'] = df['gender'].map({1:0, 2:1})

# Compute BMI (height in cm -> m)
if {'height','weight'}.issubset(df.columns):
    df['height_m'] = df['height'] / 100.0
    df['bmi'] = df['weight'] / (df['height_m'] ** 2)
    # Remove impossible BMI values (e.g., based on medical plausibility)
    df = df[(df['bmi'] > 10) & (df['bmi'] < 60)]

# Age bins (optional, useful for interpretability or specific models)
# We might drop this later if keeping the continuous 'age' feature
if 'age' in df.columns:
    df['age_bin'] = pd.cut(df['age'], bins=[0,34,44,54,64,120], labels=['<35','35-44','45-54','55-64','65+'])

# Quick final info before EDA
print("Shape after feature engineering:", df.shape)
display(df[['age','gender','height','weight','bmi','ap_hi','ap_lo','cholesterol','gluc','smoke','alco','active','cardio']].describe(include='all'))


# 2. EDA: distributions and relationships

plt.figure(figsize=(6,4))
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Age distribution')
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(df['bmi'].dropna(), bins=30, kde=True)
plt.title('BMI distribution')
plt.show()

# BP vs target
plt.figure(figsize=(8,4))
sns.boxplot(x='cardio', y='ap_hi', data=df)
plt.title('Systolic BP by CVD target')
plt.show()

# CVD rate by cholesterol level
# Calculate proportion directly for clarity
ct = df.groupby('cholesterol')['cardio'].mean().reset_index()
plt.figure(figsize=(6,4))
sns.barplot(x='cholesterol', y='cardio', data=ct)
plt.ylabel('Proportion with CVD')
plt.title('CVD rate by cholesterol level')
plt.show()

# Correlation heatmap for numeric features
num_cols = ['age','bmi','ap_hi','ap_lo','height','weight']
plt.figure(figsize=(8,6))
sns.heatmap(df[num_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation matrix (numeric)')
plt.show()

# ---- Modification Starts Here ----

# 3. Final Feature Selection
# Drop columns that are redundant or were intermediate calculations
# 'height_m' was used to calculate BMI. BMI is kept, height_m can go.
# 'age_bin' is a categorized version of 'age'. Keeping continuous 'age'
# is generally better for models, so drop 'age_bin'.
# We keep 'height' and 'weight' along with 'bmi' as they might contain
# independent predictive information.

columns_to_drop = ['height_m', 'age_bin']
final_df = df.drop(columns=columns_to_drop)

# ---- Modification Ends Here ----


# Save cleaned and final selected data
final_df.to_csv('train.csv', index=False)


print(f"Cleaned data with selected features saved to train.csv")
print(f"Shape: {final_df.shape}")
print(f"Columns: {list(final_df.columns)}")
print(f"First 5 rows:\n")
display(final_df.head())
```

Features (10)

- age - Age in years  
- gender - Gender (0=female, 1=male)  
- ap_hi - Systolic blood pressure  
- ap_lo - Diastolic blood pressure  
- cholesterol - Cholesterol level  
- gluc - Glucose level  
- smoke - Smoking status  
- alco - Alcohol intake  
- active - Physical activity  
- bmi - Body Mass Index (calculated feature)  

Target (1)

- cardio - Cardiovascular disease (0=No, 1=Yes)
